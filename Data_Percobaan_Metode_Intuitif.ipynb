```python
import numpy as np
import pandas as pd
from google.colab import files

# Fungsi untuk mengelompokkan data berdasarkan label klaster
def cluster_data_by_label(data, label_column):
    unique_labels = data[label_column].unique()
    clusters = []

    for cluster_label in unique_labels:
        cluster_points = data[data[label_column] == cluster_label][['latitude', 'longitude', label_column]]
        center = cluster_points[['latitude', 'longitude']].mean().values
        points = cluster_points.values.tolist()
        cluster = {'center': center, 'points': points}
        clusters.append(cluster)

    return clusters

# Menentukan kolom label klaster (gantilah 'label_cluster' dengan nama kolom label klaster Anda)
label_column = 'label_cluster'

# Mengelompokkan data berdasarkan label klaster dan menghitung centroid
result_clusters = cluster_data_by_label(data, label_column)
print(result_clusters)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from math import radians, sin, cos, sqrt, atan2

def haversine_distance(coord1, coord2):
    # Extract latitude and longitude from the coordinates
    lat1, lon1 = coord1
    lat2, lon2 = coord2

    # Konversi koordinat dari derajat ke radian
    lat1, lon1, lat2, lon2 = map(radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    distance = 6371 * c  # Radius of Earth in kilometers

    return distance


# Fungsi untuk menghitung matriks jarak Haversine antar lokasi
# Fungsi untuk menghitung matriks jarak Haversine antar lokasi
def generate_distance_matrix(locations):
    n = len(locations)

    # membuat matriks nol dengan ukuran n x n
    distances = np.zeros((n, n))

    # Membangun jarak antar lokasi berdasarkan fungsi haversine distance
    for i in range(n):
        for j in range(n):
            distances[i][j] = haversine_distance(locations[i], locations[j])

    return distances

distances = generate_distance_matrix(cluster_centers)

# Input
depot_location = (-6.194515855159909, 106.80531792027882)  # Koordinat depot
```

```python
# Fungsi untuk inisialisasi kromosom dengan urutan titik menggunakan heuristik ANN
def ann_initialization(n, start_point, start_time, end_time, result_clusters, cluster_centers):
    remaining_points = set(range(1, n + 1))  # set lokasi yang mungkin
    chromosome = [0] + [start_point + 1] + [0]  # Menambahkan depot(0) di depan dan belakang (0)

    # jika titik awal masih ada di dalam remaining_points
    if start_point + 1 in remaining_points:
        # maka akan dihapus dari remaining_points
        remaining_points.remove(start_point + 1)

    current_time = start_time  # Inisiasi current_time sebagai Waktu awal

    # Loop pembentukan rute menggunakan ANN
    # Loop akan terus berjalan selama masih ada titik yang harus dikunjungi
    while remaining_points:

        # Titik terakhir yang telah ditambahkan ke rute sebelumnya
        current_point = chromosome[-2] - 1  # Perbarui indeks dengan current_point - 1

        # Mengecek apakah titik terakhir yang ditambahkan masih di dalam rentang yang valid
        if 0 <= current_point < n:

            # Pilih titik terdekat berdasarkan jarak haversine
            # x adalah variabel yang mewakili setiap elemen dalam remaining_point
            # centers[current_point] adalah titik saat ini
            # [x - 1] karena indeks dalam python dimulai dari nol
            nearest_neighbor = min(remaining_points, key=lambda x: haversine_distance(cluster_centers[current_point], cluster_centers[x - 1]))

            # Perhitungan waktu layanan drone
            # Mendapatkan informasi terkait cluster dari titik saat ini
            cluster_sizes = {tuple(np.round(tuple(center[:2]), decimals=3)): len(cluster['points']) for center, cluster in zip(cluster_centers, result_clusters)}

            # Perhitungan waktu tempuh truk
            truck_travel_time = haversine_distance(cluster_centers[current_point], cluster_centers[nearest_neighbor - 1]) * 2

            # Mendapatkan informasi terkait cluster dari titik saat ini
            current_cluster_size = cluster_sizes.get(current_point, 0)
            # Menghitung service time pada titik terakhir sebelum ke depot
            service_time_drone = (2 * 0.57 * 4) + (3 * current_cluster_size)+5

            # Perhitungan total waktu untuk jalur
            # Cluster size diperoleh dari informasi cluster terkait
            current_cluster_size = cluster_sizes.get(current_point, 0)
            total_time = current_time +  service_time_drone + truck_travel_time

            # Memastikan bahwa total waktu berada dalam time windows
            if start_time <= total_time <= end_time:
                # jika memenuhi batasan waktu maka akan memasukkan nearest_neighbor ke dalam kromosom
                # -1 artinya akan dimasukkan ke dalam posisi sebelum elemen terakhir, dalam hal ini elemen terakhir harus diisi oleh depot
                chromosome.insert(-1, nearest_neighbor)
                # Menghapus titik yang sebelumnya dimasukkan ke dalam kromosom dari remaining_point
                remaining_points.remove(nearest_neighbor)
                # Waktu diperbarui dengan nilai total_time
                current_time = total_time
        else:
            # Jika sudah diluar rentang yang valid, loop dihentikan
            break

    return chromosome

# Membangkitkan populasi awal dengan urutan titik menggunakan heuristik ANN
def generate_initial_population_ann(n, size, start_time, end_time):
    population = []
    for i in range(size):
        # Inisialisasi kromosom dengan urutan titik menggunakan heuristik ANN
        chromosome = ann_initialization(n, i, start_time, end_time, result_clusters, cluster_centers)
        # Kromosom yang terbentuk dimasukkan dalam populasi
        population.append(chromosome)
    return population

# Fungsi untuk menghitung time windows
def calculate_time_windows(chromosome, start_time, distances, depot_location, cluster_sizes):
    current_time = start_time  # Inisiasi current_time sebagai titik awal
    time_windows = []  # variabel untuk menyimpan time_windows

    # Membuat variabel baru untuk menyimpan informasi jumlah titik di setiap pusat kluster
    cluster_sizes = {tuple(np.round(tuple(center[:2]), decimals=3)): len(cluster['points']) for center, cluster in zip(cluster_centers, result_clusters)}

    # untuk memanipulasi agar pasangan titiknya berturut-turut (misalkan: 1 (x1,y1) ke 2 (x2,y2))
    for i in range(len(chromosome) - 1):
        # current_point diatur ke nilai dari elemen kromosom i
        current_point = chromosome[i]
        # next_point diatur ke nilai pada elemen kromosom i + 1
        next_point = chromosome[i + 1]

        # Memeriksa apakah current_point merupakan titik depot
        if current_point == 0:
            # Jika current_point merupakan depot maka akan dihitung time windows dari depot
            # menghitung jarak dan time windows dari depot ke titik awal
            depot_to_start_distance = haversine_distance(depot_location, cluster_centers[next_point - 1])
            # 2 disini adalah kecepatan rata-rata truk dalam menit/km
            truck_travel_time = depot_to_start_distance * 2
            total_time = current_time + truck_travel_time

        # Memeriksa apakah next_point adalah depot
        elif next_point == 0:
            # Jika next_point merupakan depot maka akan dihitung time windows dari titik terakhir ke depot
            # Menghitung jarak dan time_windows dari titik terakhir ke depot
            end_to_depot_distance = haversine_distance(cluster_centers[current_point - 1], depot_location)

            # 2 disini adalah kecepatan rata-rata truk dalam menit/km
            truck_travel_time = end_to_depot_distance * 2

            current_cluster_size = cluster_sizes.get(tuple(np.round(tuple(cluster_centers[current_point - 1][:2]), decimals=3)), 0)
            # Menghitung service time pada titik terakhir sebelum ke depot
            service_time_drone = (2 * 0.57 * 4) + (3 * current_cluster_size) + 5

            total_time = current_time + service_time_drone + truck_travel_time

        # jika current_point bukan merupakan depot dan titik terakhir
        else:

            current_cluster_size = cluster_sizes.get(tuple(np.round(tuple(cluster_centers[current_point - 1][:2]), decimals=3)), 0)
            # Menghitung service time drone
            service_time_drone = (2 * 0.57 * 4) + (3 * current_cluster_size) + 5

            # Menghitung waktu tempuh truk
            # 2 adalah kecepatan rata-rata truk dalam menit/km
            truck_travel_time = distances[current_point - 1][next_point - 1] * 2

            # Menghitung total waktu
            total_time = current_time + service_time_drone + truck_travel_time

        # untuk menyimpan informasi waktu dan membentuk jendela waktu untuk setiap segmen rute dalam kromosom.
        time_windows.append((convert_to_time(int(current_time)), convert_to_time(int(total_time))))
        # perbarui current_time dengan total time
        current_time = total_time

    return time_windows

#fungsi untuk menghitung total jarak

def calculate_total_distance(chromosome, distances, depot_location, start_time):
    total_distance = 0  # Inisiasi total jarak tempuh rute
    current_time = start_time  # Inisiasi current_time sebagai waktu awal

    for i in range(len(chromosome) - 1):
        current_point = chromosome[i]
        next_point = chromosome[i + 1]

        if current_point == 0 and next_point != 0:
            depot_to_start_distance = haversine_distance(depot_location, cluster_centers[next_point - 1])
            total_distance += depot_to_start_distance
            truck_travel_distance = depot_to_start_distance

        elif current_point != 0 and next_point == 0:
            end_to_depot_distance = haversine_distance(cluster_centers[current_point - 1], depot_location)
            total_distance += end_to_depot_distance
            truck_travel_distance = end_to_depot_distance
        else:
            truck_travel_distance = distances[current_point - 1][next_point - 1]
            total_distance += truck_travel_distance

    return total_distance

# Fungsi untuk mengonversi waktu dari menit ke format jam
def convert_to_time(minutes):
    hours = int(minutes) // 60
    minutes = int(minutes) % 60
    return f"{hours:02d}:{minutes:02d}"

# Input
depot_location = (-6.194515855159909, 106.80531792027882)  # Koordinat depot
# Input untuk memperoleh populasi awal
n = len(cluster_centers)
ukuran_populasi = 19
start_time = 480
end_time = 1020

# Membuat matriks jarak
distances = generate_distance_matrix(cluster_centers)

# Membangkitkan populasi menggunakan ANN
population = generate_initial_population_ann(n, ukuran_populasi, start_time, end_time)

# Menghasilkan output
for idx, chrom in enumerate(population):
    total_distance = calculate_total_distance(chrom, distances, depot_location, start_time)
    tw = calculate_time_windows(chrom, start_time, distances, depot_location, cluster_sizes)
    tw_in_hours = [(start, end) for start, end in tw]

    print(f"Rute {idx + 1} : {chrom},\nTime Windows: {tw_in_hours}, \nTotal Distance: {total_distance} km")
```

```python
import random
# Fungsi untuk menghitung nilai fitness
def calculate_fitness(chromosome, distances, depot_location, start_time, end_time):
    total_distance = 0 #Inisiasi total jarak tempuh rute

    # untuk memanipulasi agar pasangan titiknya berturut-turut (misalkan : 1 (x1,y1) ke 2 (x2,y2))
    for i in range(len(chromosome) - 1):
        # current_point diatur ke nilai dari elemen kromosom i
        current_point = chromosome[i]
        # next_poit diatur ke nilai pada elemen kromosom i + 1
        next_point = chromosome[i + 1]

        # Jika current_point merupakan depot maka akan dihitung jarak dan time windows dari depot
        # menghitung jarak dan time windpws dari depot ke titik awal
        if current_point == 0 and next_point != 0:
            # Jarak dan waktu tempuh dari depot ke lokasi awal
            depot_to_start_distance = haversine_distance(depot_location, cluster_centers[next_point - 1])
            # 2 disini adalah kecepatan rata-rata truk dalam menit/km
            truck_travel_time = depot_to_start_distance * 2
            # menghitung jarak dari depot ke titik awal
            total_distance += depot_to_start_distance

        # Memeriksa apakah next_point adalah depot
        elif current_point != 0 and next_point == 0:
            # Jika next_point merupakan depot maka akan dihitung time windows dari titik terakhir ke depot
            # Jarak dan waktu tempuh dari lokasi akhir ke depot
            end_to_depot_distance = haversine_distance(cluster_centers[current_point - 1], depot_location)
            # 2 disini adalah kecepatan rata-rata truk dalam menit/k
            truck_travel_time = end_to_depot_distance * 2
            # menghitung jarak dari titik terakhir ke depot
            total_distance += end_to_depot_distance

        else:
            # Perhitungan waktu tempuh truk
            truck_travel_time = distances[current_point - 1][next_point - 1] * 2

            # Perhitungan total waktu dan total distance untuk jalur
            total_distance += distances[current_point - 1][next_point - 1]

    # Menghitung nilai fitness
    fitness = (1 / (total_distance ))

    return fitness

# Fungsi seleksi menggunakan Roulette Wheel Selection
def roulette_wheel_selection(population, fitness_values):
    if not fitness_values:
        return random.randint(0, len(population) - 1)

    total_fitness = sum(fitness_values)
    if total_fitness == 0:
        # If total_fitness is zero, set all fitness values to 1 to avoid division by zero
        fitness_values = [1] * len(fitness_values)
        total_fitness = len(fitness_values)

    pick = random.uniform(0, total_fitness)
    current_fitness = 0

    for i in range(len(population)):
        current_fitness += fitness_values[i]
        if current_fitness > pick:
            return i  # Return the index of the selected individual

# Fungsi crossover menggunakan Order Crossover (OX)
def order_crossover(parent1, parent2):
    # Rentang yang akan diambil dari parent1 dan parent2 untuk disalin ke anak 1 dan 2
    start_index = 1
    end_index = 19

    child1 = [0] * len(parent1)  # child diinisialisasi sebagai array dengan panjang yang sama dengan panjang kromosom orang tua
    child2 = [0] * len(parent1)

    child1[start_index:end_index + 1] = parent1[start_index:end_index + 1]  # Menyalin sub kromosom dari parent1
    child2[start_index:end_index + 1] = parent2[start_index:end_index + 1]

    index_child1 = end_index + 1  # sebagai posisi yang akan diisi oleh parent 2
    index_child2 = end_index + 1

    # mengisi gen pada child yang tidak disalin oleh parent 1
    for gene in parent2[end_index + 1:] + parent2[:end_index + 1]:
        if gene not in child1:
            child1[index_child1 % len(parent1)] = gene
            index_child1 += 1

    for gene in parent1[end_index + 1:] + parent1[:end_index + 1]:
        if gene not in child2:
            child2[index_child2 % len(parent1)] = gene
            index_child2 += 1

    return child1, child2


# Fungsi mutasi menggunakan Inversion Mutation
def inversion_mutation(chromosome, prob_mutasi):
    # Rentang kromosom yang akan diambil untuk melakukan mutasi
    start_index = 1
    end_index = len(chromosome) - 2

    for i in range (len(chromosome)):
      if random.random()<prob_mutasi :
         #untuk memilih dua posisi secara acak
          pos1, pos2 = sorted(random.sample(range(start_index, end_index + 1), 2))
          # Melakukan inversi (memutar rute dari dua posisi acak yang telah dipilih)
          chromosome[pos1:pos2+1] = list(reversed(chromosome[pos1:pos2+1]))
    return chromosome

def genetic_algorithm(population_size, chromosome_length, generations, prob_crossover, prob_mutation, distances, depot_location, start_time, end_time):
    population = generate_initial_population_ann(chromosome_length, population_size, start_time, end_time)

    for generation in range(generations):
        gabungan = []
        child = []
        fitnesses = []

        for _ in range(population_size // 2):
            idxParent1 = roulette_wheel_selection(population, fitnesses)
            idxParent2 = roulette_wheel_selection(population, fitnesses)
            parent1 = population[idxParent1][:]
            parent2 = population[idxParent2][:]

            # Crossover
            if random.uniform(0, 1) < prob_crossover:
                child1, child2 = order_crossover(parent1, parent2)
            else:
                child1, child2 = parent1[:], parent2[:]

            # Mutasi
            child1 = inversion_mutation(child1, prob_mutation)
            child2 = inversion_mutation(child2, prob_mutation)

            child.extend([child1, child2])

        gabungan = population + child

        for i in range(len(gabungan)):
            fitnesses.append(calculate_fitness(gabungan[i], distances, depot_location, start_time, end_time))

        mergedArray = [(gabungan[i], fitnesses[i]) for i in range(len(gabungan))]
        mergedArray = sorted(mergedArray, key=lambda a: a[1], reverse=True)

        pop = []
        tmp = []
        t = 0
        jumlah = 0

        for i in range(len(gabungan)):
            if i >= 1:
                nilai = calculate_fitness(gabungan[i], distances, depot_location, start_time, end_time)
                if nilai == mergedArray[jumlah - 1][1]:
                    tmp.append(mergedArray[i][0])
                    t += 1
                else:
                    pop.append(mergedArray[i][0])
                    jumlah += 1
            else:
                pop.append(mergedArray[i][0])
                jumlah += 1

            if jumlah == population_size:
                break

        z = 0

        while jumlah < population_size:
            pop.append(tmp[z])
            z += 1
            jumlah += 1

        population = pop

    return population

# Penggunaan algoritma genetika
population_size = 19
n = 19
generations = 100
prob_crossover = 0.6
prob_mutasi = 0.001

# Contoh penggunaan
pop = generate_initial_population_ann(n, population_size, start_time, end_time)
pop = list(pop)
pop = genetic_algorithm(population_size, n, generations, prob_crossover, prob_mutasi, distances, depot_location, start_time, end_time)
print("Generasi ke- " + str(generations))
print("Rute terbaik : " + str(pop[0]))
print("Fitness : " + str(calculate_fitness(pop[0], distances, depot_location, start_time, end_time)))
```
